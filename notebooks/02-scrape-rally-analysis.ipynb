# 02 rally-analysis scraping + processing example

```python
import pandas as pd

from infotennis.scrapers.scrape_match_data import scrape_ATP_match_data
from infotennis.processing.processing_rallys import process_rally_analysis
```

```python
raw_data = scrape_ATP_match_data(2023, 352, "ms005", "rally-analysis")
raw_data
```

```python
pd.DataFrame(raw_data['rallyData'])
```

```python
df_rally_pts = process_rally_analysis(2023, 352, "ms005", "Quarter-Final", raw_data)
df_rally_pts
```

#### Quick Analysis

How many points in the 1-3 shots range did each player win/lose?

```python
df_rally_pts[df_rally_pts.shot_number.isin(['1','2','3']) ].groupby(['player_id'])['outcome'].value_counts()
```
